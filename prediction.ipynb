```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

# Loading All Trained Models

```python
model = load_model('artifacts/model.h5')

with open('artifacts/encoder.pkl','rb') as file:
    encoder = pickle.load(file)

with open('artifacts/scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
## Example data
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
input_data_df = pd.DataFrame([input_data])
input_data_df.head()
```

```python
encoded = encoder.transform([[input_data['Gender'],input_data['Geography']]])

encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['Gender','Geography']))
```

```python
encoded_df.head()
```

```python
input_data_df = pd.concat([input_data_df.drop(['Gender','Geography'],axis=1),encoded_df],axis=1)
```

```python
input_data_df.head()
```

```python
input_scaled=scaler.transform(input_data_df)
```

```python
input_scaled
```

```python
##Predict Churn
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba>0.5:
    print("Th customer is likely to churn")

print('Customer is not likely to churn')
```
