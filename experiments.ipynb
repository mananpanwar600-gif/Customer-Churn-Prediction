```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
```

# Load the Dataset

```python
data = pd.read_csv('dataset/Churn_Modelling.csv')
data.head()
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

# Encoding

```python
encoder = OneHotEncoder(drop=None,sparse_output = False)
encoded_df=encoder.fit_transform(data[['Gender', 'Geography']])
```

```python
encoded_df = pd.DataFrame(encoded_df,columns=encoder.get_feature_names_out(['Gender','Geography']))
encoded_df
```

```python
data = pd.concat([data.drop(['Geography','Gender'],axis=1),encoded_df],axis=1)
```

```python
data.head()
```

```python
## Save this encoding in pickle file
with open('artifacts/encoder.pkl','wb') as file:
    pickle.dump(encoder,file)
```

# Independent and Dependent Features

```python
X = data.drop('Exited',axis=1)
y=data['Exited']
```

```python
X.head()
```

```python
y.head()
```

# Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

# Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
## Save this scaler in pickle
with open('artifacts/scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

# ANN 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
## Build ANN Model
model = Sequential([
            Input(shape=(X_train.shape[1],)),
            Dense(64,activation='relu'), ##HL 1
            Dense(32,activation='relu'),  ##HL 2
            Dense(1,activation='sigmoid') ##Output
])
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
```

```python
## Compile model
model.compile(optimizer=opt,loss=loss,metrics=["accuracy"])
```

```python
## Set up the TensorBoard
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## Set up Early Stopping
early_stopping_callback = EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
## Training Model
history = model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('artifacts/model.h5')
```
